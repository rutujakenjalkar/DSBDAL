```python
#DETECTING OUTLIERS THROUGH ZSCORE

mean=np.mean(data["PERCENATAGE"])
std=np.std(data['PERCENATAGE'])
print("the mean is:",mean)
print("the std is:",std)
```

```python
Q1=data['PERCENATAGE'].quantile(0.25)
Q2=data['PERCENATAGE'].quantile(0.50)
Q3=data['PERCENATAGE'].quantile(0.75)

IQR=Q3-Q1

lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR

print("the lower limit:",lower_limit)
print("the upper limit:",upper_limit)

```

```python
outliers=[]
for i in data['PERCENATAGE']:
    if i>upper_limit or i<lower_limit:
        outliers.append(i)
        
print("the outliers are",outliers)
```
