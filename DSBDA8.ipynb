```python
#IMPORTING ALL THE NECESSARY DEPENDENCIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#LOAD THE GIVEN DATA SET
data=sns.load_dataset('titanic')
data
```

```python
sns.distplot(data['fare'])
```

```python
#GET RID OF THE LINE KDE=FALSE 
sns.distplot(data['fare'],kde=False)
```

```python
#data split in 10 bins for representaiton on the graph
sns.displot(data['fare'],kde=False,bins=10)
```

```python
#CREATING JOINT PLOTS
sns.jointplot(x=data['age'],y=data['fare'],data=data)
```

```python
#TO CHANGE THE STYLE OF THE DOTS kind property is used ie set to hex
sns.jointplot(x=data['age'],y=data['fare'],data=data,kind='hex')
```

```python
#PLOTTING A BAR PLOT
sns.barplot(x=data['sex'],y=data['fare'],data=data)
```

```python
#A rug plot is a simple way to visualize the distribution of a dataset by drawing small
#vertical lines (or ticks) along an axis â€” one for each data point.projections of points on one of the  axis .default=x
```

```python
sns.rugplot(data['fare'])
```

```python
#COUNTPLOT -GIVES THE COUNT OF EACH CATEOGORY OF ONE PARTICULAR CATOEGORY
```

```python
sns.countplot(data['sex'],data=data)
```

```python
#A box plot is a graphical summary of a dataset that shows its median, quartiles, range
#, and outliers, helping to visualize the distribution and variability.
```

```python
sns.boxplot(x=data['sex'],y=data['age'],data=data)
```

```python
#VIOLIN PLOT
#A violin plot combines a box plot and a KDE (Kernel Density Estimate) plot to show the distribution,
#central tendency, and probability density of a dataset in a single visual.
```

```python
sns.violinplot(x=data['sex'],y=data['age'],data=data)
```

```python
sns.violinplot(x=data['sex'],y=data['fare'],data=data)
```

```python
#STRIP PLOT
sns.stripplot(x=data['sex'],y=data['age'],data=data)
```

```python
sns.stripplot(x=data['sex'],y=data['fare'],data=data)
```

```python
#SWARMPLOT
sns.swarmplot(x=data['sex'],y=data['age'],data=data)
```
