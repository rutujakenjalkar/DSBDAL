```python
Q1=data['medv'].quantile(0.25)
Q3=data['medv'].quantile(0.75)

IQR=Q3-Q1
lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
print("lower limit:",lower_limit)
print("upper limit",upper_limit)
```

```python
medv_outliers=[]
for i in data['medv']:
    if i>upper_limit or i<lower_limit:
        medv_outliers.append(i)
        
print("medv outliers are:",medv_outliers)
```

```python
data[data['medv']<lower_limit].index
```

```python
#REMOVING THE OUTLIERS 

outliers_indices=data[(data['medv']<lower_limit)|(data['medv']>upper_limit)].index

df1=data.drop(outliers_indices)
```

```python
df1
```

```python
df1.boxplot()
```

```python
#PREPARING THE DATA FOR THE MODEL
```

```python
X=df1.drop(['medv'],axis=1)
Y=df1['medv']
```

```python
print("training set:")
print(X)
```

```python
print("testing set:")
print(Y)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
import sklearn 
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
```

```python
xtrain
```

```python
ytrain
```

```python
xtrain.fillna(0, inplace=True)
ytrain.fillna(0, inplace=True)
xtest.fillna(0,inplace=True)
```

```python
model = lm.fit(xtrain, ytrain)
```

```python
ytrain_pred=lm.predict(xtrain)
ytest_pred=lm.predict(xtest)
```

```python
ytrain
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error 
```

```python
mae=mean_absolute_error(ytrain,ytrain_pred)
print("mae:",mae)
```

```python
mae=mean_absolute_error(ytest,ytest_pred)
print("mae:",mae)
```

```python
#MEAN SQUARED ERROR 
```

```python
mse_train=mean_squared_error(ytrain,ytrain_pred)
print("mse for train set:",mse_train)
```

```python
mse_test=mean_squared_error(ytest,ytest_pred)
print("mse for test set:",mse_test)
```

```python
rmse_train=np.sqrt(mean_squared_error(ytrain,ytrain_pred))
print("rmse train set:",rmse_train)
```

```python
#R2 Score
r2_train=r2_score(ytrain,ytrain_pred)
print("r2 score for the train set:",r2_train)
```

```python
r2_test=r2_score(ytest,ytest_pred)
print("r2 score for the train set:",r2_train)
```

```python
r2_test=r2_score(ytest,ytest_pred)
print("r2 score for the test set:",r2_test)
```
